```python
import itertools
import math
import matplotlib.pyplot as plt



# This is a dictionary that maps amino acids (keys) to their hydropathy values (values) calculated using the Kyte Doolittle
# method. These values are common so can easily be found on the internet. 

amino_acids = {'I': 4.5, 'F': 2.8, 'V': 4.2, 'L': 3.8, 'W': -0.9,
               'M': 1.9, 'A': 1.8, 'G': -0.4, 'C': 2.5, 'Y': -1.3,
               'P': -1.6, 'T': -0.7, 'S': -0.8, 'H': -3.2, 'E': -3.5,
               'N': -3.5, 'Q': -3.5, 'D': -3.5, 'K': -3.9, 'R': -4.5}

#The user can enter an amino acid sequence of their choice and desired length and a window size. 

sequence = input('Enter a sequence:')
n = input('Enter a window size:')
window_size = int(n)
x = 0
y = window_size
middleElement = math.floor(y/2) #Using floor we round' 'down' everytime. 
#print(middleElement)


summed_hydropathy = []
tempList = []

for i in range(0,len(sequence)-window_size+1):      #To loop through each sequence within the window size
    #print(sequence[x:y])
    #print(sequence[x+middleElement])
    j = i
    counter = 0             
    avgVal = 0
    while counter != window_size:               # Sum up all the values within the window frame
        avgVal += amino_acids[sequence[j]]  
        counter +=1
        j +=1
    avgVal /= window_size           # Just calculates the average value
    tempString = str(sequence[x+middleElement])+':'+str(avgVal)         # Just for visual purposes; Displays the middle element and the average value that should correspond with it
    tempList.append(tempString)
    hydropathyString = avgVal
    summed_hydropathy.append(hydropathyString)
    x +=1
    y+=1

#print(*tempList, sep='\n')  
#print(summed_hydropathy, sep= '\n')
 
sequence = list(sequence) # converts the sequence to a list 


a = int(window_size/2) + 1
b = len(sequence)-a +2
pos = range(a,b)

position = []
for a in pos:
    position.append(a)
    
#print(position)

#print(len(summed_hydropathy)) To ensure the lenghs of the two lists that will be plotted against each otehr is the same
#print(len(position))

plt.plot(position,summed_hydropathy,'-b', label='Hydropathy Plot')


plt.xlabel('Amino Acid Position', size = 12)
plt.ylabel('Hydrophobicity', size = 12)
plt.grid(linestyle = '--')

ax = plt.gca()

#ax.xaxis.set_major_locator(MaxNLocator(integer=True))


plt.show()

    
```
